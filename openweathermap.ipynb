```python
import requests
import pandas as pd
api_key="e1ef1f5ee6503c212ec04e9c54573740"
url2=f"http://api.openweathermap.org/data/2.5/air_pollution/forecast?lat={lat}&lon={lon}&appid={api_key}"
respond=requests.get(url2)
cord1=respond.json()
```

```python
cord1["list"]
```

```python
d={"aqi":[],"No":[],"So2":[],"pm10":[]}
for i in cord1["list"]:
    d["aqi"].append(i["main"]["aqi"])
    d["No"].append(i["components"]["no"])
    d["So2"].append(i["components"]["so2"])
    d["pm10"].append(i["components"]["pm10"]) 
df1=pd.DataFrame(d)
print(df1)
```

```python
d={"date":[],"temperature":[],"pressure":[],"humidity":[],"visibility":[],"desc":[]}
for i in cord["list"]:
    d["date"].append(i["dt_txt"])
    d["temperature"].append(i["main"]["temp"])    
    d["pressure"].append(i["main"]["pressure"])    
    d["humidity"].append(i["main"]["humidity"])
    d["visibility"].append(i["visibility"])
    d["desc"].append(i["weather"][0]["description"])   
df=pd.DataFrame(d)
print(df)
```

```python
from bs4 import BeautifulSoup
url="https://subslikescript.com/movie/Titanic-120338"
responds=requests.get(url).text
soup=BeautifulSoup(responds,"html.parser")
box=soup.find("article",class_="main-article")
title=box.find('h1').text
print(title)
script=box.find("div",class_="full-script").text
print(script)
```

```python
f=open(f"{title}.txt","w",encoding="utf-8")
f.write(script)
f.close()
```

```python
url2="https://subslikescript.com/movies"
responds=requests.get(url2).text
soup=BeautifulSoup(responds,"html.parser")
box=soup.find("article",class_="main-article")
print(box.findAll("a",href=True))
```

```python
links=[]
for i in box.findAll("a",href=True):
    links.append(i["href"])
print(links)
```

```python
for i in links:
    try:
        url="https://subslikescript.com"+i
        responds=requests.get(url).text
        soup=BeautifulSoup(responds,"html.parser")
        box=soup.find("article",class_="main-article")
        title=box.find('h1').text
        script=box.find("div",class_="full-script").text
        f=open(f"{title}.txt","w",encoding="utf-8")
        f.write(script)
        f.close()
    except:
        print()
```
